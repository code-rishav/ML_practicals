```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.decomposition import PCA
from sklearn.cluster import KMeans

# Load the iris dataset
iris = load_iris()
X, y = iris.data, iris.target

# Apply PCA to select the best 2 features
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

# Plot the scatter graph
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=y, cmap='viridis', edgecolor='k')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('Scatter Plot of Iris Dataset after PCA')
plt.colorbar(label='Species')
plt.grid(True)
plt.show()

# Apply k-means clustering algorithm
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X_pca)

# Plot the clusters
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], c=kmeans.labels_, cmap='viridis', edgecolor='k')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=200, c='red', marker='X', label='Cluster Centers')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('K-Means Clustering of Iris Dataset after PCA')
plt.colorbar(label='Cluster')
plt.legend()
plt.grid(True)
plt.show()
```
